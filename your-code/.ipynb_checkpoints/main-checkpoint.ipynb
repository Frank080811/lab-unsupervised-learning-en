# Challenge 3 - Data Preprocessing

One problem with the dataset is the value ranges are remarkably different across various categories (e.g. `Fresh` and `Grocery` compared to `Detergents_Paper` and `Delicassen`). If you made this observation in the first challenge, you've done a great job! This means you not only completed the bonus questions in the previous Supervised Learning lab but also researched deep into [*feature scaling*](https://en.wikipedia.org/wiki/Feature_scaling). Keep on the good work!

Diverse value ranges in different features could cause issues in our clustering. The way to reduce the problem is through feature scaling. We'll use this technique again with this dataset.

#### We will use the `StandardScaler` from `sklearn.preprocessing` and scale our data. Read more about `StandardScaler` [here](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html#sklearn.preprocessing.StandardScaler).

*After scaling your data, assign the transformed data to a new variable `customers_scale`.*

```python
# Encode Categorical Values

df_encoded = pd.get_dummies(wholesale_df_log, columns=["Channel", "Region"], drop_first=True)
```

```python
# Your import here:

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_encoded[num_cols] = scaler.fit_transform(df_encoded[num_cols])

```

```python
df_clean = df_encoded.copy()
df_clean.head()
```

These steps successfully prepare the dataset for machine learning by making the features both numerical and comparable in scale.

First, the categorical variables Channel and Region were converted into dummy variables using one-hot encoding. This allows models to use these categories without misinterpreting them as numeric values. The drop_first=True option avoids multicollinearity by removing one category from each group and treating it as the baseline.

Next, the numerical spending features were standardized using StandardScaler. This transformation centers each feature around zero and scales it to have unit variance. The resulting values represent how far above or below the average each observation lies. Positive values indicate higher-than-average spending, while negative values reflect lower-than-average spending.

The resulting cleaned dataset shows well-scaled numerical values alongside boolean indicators for category membership, making it ideal for clustering, PCA, or other machine-learning algorithms that rely on distance or variance.

# Challenge 4 - Data Clustering with K-Means

Now let's cluster the data with K-Means first. Initiate the K-Means model, then fit your scaled data. In the data returned from the `.fit` method, there is an attribute called `labels_` which is the cluster number assigned to each data record. What you can do is to assign these labels back to `customers` in a new column called `customers['labels']`. Then you'll see the cluster results of the original data.

```python
from sklearn.cluster import KMeans

# Selecting the features for clustering (all numeric columns)
X = df_clean.copy()

# Initialize K-Means model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model
kmeans.fit(X)

# Assign labels back to the original data
df_clean['labels'] = kmeans.labels_

# View the first few results
df_clean.head()

```

### Looking to the elbow we can choose 2 like the correct number of clusters

```python
from sklearn.cluster import KMeans

# Fit the K-Means model with k=2 on the scaled dataset
kmeans_2 = KMeans(n_clusters=2, random_state=42).fit(df_clean)

# Predict cluster labels
labels = kmeans_2.predict(df_clean)

# Convert to list
clusters = labels.tolist()

# Assign labels back to dataframe
df_clean["cluster"] = clusters

df_clean.head()
```

These steps applied K-Means to group customers based on their spending habits. After scaling the data, the algorithm identified two clear customer groups. Most customers in this sample fall into Cluster 0, meaning their spending patterns are fairly similar to each other. One customer (row 3) stands out and is placed in Cluster 1, likely because their buying behavior differs noticeably, shown by the negative value in Milk and unusually high Frozen spending.

In simple terms, the clustering worked: it separated customers into two meaningful segments based on how they shop. This can help businesses understand different types of buyers and tailor strategies to each group.

Count the values in `labels`.

```python
df_clean["labels"].value_counts()
```

These results show that your customers naturally fall into three different groups. The biggest group is Cluster 1, with 202 people, meaning this is the most common spending style in your dataset. Cluster 0 is also fairly large, with 170 customers who shop in a noticeably different way. Cluster 2 is much smaller, with just 68 customers, suggesting this group has a more unusual or distinctive buying pattern.

In simple terms, most customers behave similarly, but there’s a small segment that stands out from the rest.

# Challenge 5 - Data Clustering with DBSCAN

Now let's cluster the data using DBSCAN. Use `DBSCAN(eps=0.5)` to initiate the model, then fit your scaled data. In the data returned from the `.fit` method, assign the `labels_` back to `customers['labels_DBSCAN']`. Now your original data have two labels, one from K-Means and the other from DBSCAN.

```python
from sklearn.cluster import DBSCAN

# Using the same scaled data as for K-Means
X = df_clean.drop(columns=["labels", "cluster"], errors="ignore") 

# Initialiinge and fit DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan.fit(X)

# Getting DBSCAN labels
labels_dbscan = dbscan.labels_

# Assigning back to your data
df_clean["labels_DBSCAN"] = labels_dbscan

df_clean[["labels", "cluster", "labels_DBSCAN"]].head()
```

Count the values in `labels_DBSCAN`.

```python
df_clean["labels_DBSCAN"].value_counts()
```

These steps applied DBSCAN to see how a different clustering method groups the customers compared to K-Means. Unlike K-Means—which forces every customer into a cluster, DBSCAN is more selective. It only groups customers who naturally cluster together and labels unusual cases as outliers (marked as -1).

In the small sample shown, most customers that K-Means placed in Cluster 0 were flagged by DBSCAN as outliers. Only one customer was considered part of a real DBSCAN cluster. This means DBSCAN sees many of these customers as having spending patterns that don’t closely match others.

In simple terms:
👉 K-Means tries to organize everyone, even if they don't fit perfectly.
👉 DBSCAN points out who truly belongs together—and who behaves differently.

This gives you a more realistic picture of which customers naturally form groups and which ones stand out

# Challenge 6 - Compare K-Means with DBSCAN

Now we want to visually compare how K-Means and DBSCAN have clustered our data. We will create scatter plots for several columns. For each of the following column pairs, plot a scatter plot using `labels` and another using `labels_DBSCAN`. Put them side by side to compare. Which clustering algorithm makes better sense?

```python
def plot(x,y,hue):
    sns.scatterplot(x=x, 
                    y=y,
                    hue=hue)
    plt.title('Detergents Paper vs Milk ')
    return plt.show();
```

```python
# K-Means visualization
plot(
    x=df_clean["Grocery"],
    y=df_clean["Fresh"],
    hue=df_clean["labels"]      # K-Means cluster labels
)

# DBSCAN visualization
plot(
    x=df_clean["Grocery"],
    y=df_clean["Fresh"],
    hue=df_clean["labels_DBSCAN"]   # DBSCAN clustering result
)

# DBSCAN with Noise labeled
plot(
    x=df_clean["Grocery"],
    y=df_clean["Fresh"],
    hue=df_clean["labels_DBSCAN"].replace(-1, "Noise")
)
```

Let's use a groupby to see how the mean differs between the groups. Group `customers` by `labels` and `labels_DBSCAN` respectively and compute the means for all columns.

Looking at the results, K-Means clearly does a better job clustering this dataset. DBSCAN, however, doesn’t perform as well here. With the chosen settings, it only detects one small cluster, and it labels almost all other customers as outliers. That means DBSCAN doesn’t really form useful groups instead, it mostly says that customers don’t belong to any cluster.

The average values within each group confirm this: DBSCAN’s “noise” group looks very close to the overall dataset average, meaning it didn’t find structure. Meanwhile, K-Means produces several clusters with clearly different spending levels, making it far more helpful for understanding customer behavior.
In short:

* 👉 K-Means gives meaningful, interpretable clusters
* 👉 DBSCAN mostly marks customers as noise

So, K-Means is the better choice for this dataset.
